You have a dataframe `df` with the features of diamonds.

```python
# Do not make any change to this cell. Just run this cell.
import os
from seaborn import load_dataset

df = load_dataset("diamonds")
df
```

1\. (1 pt) Create a CSV file named `hw_diamonds.csv` in the `outcome` directory. The file contains 7 column names `carat`, `cut`, `color`, `clarity`, `depth`, `table`, and `price` in the first row and the values from the corresponding columns of `df` in the following rows. The delimiter is a slash (/) for columns and a new line (\n) for rows. For example, the first 3 rows in the file must look as below. Do not use Pandas.

```python
# Your answer here
if not os.path.isdir("outcome"):         
    os.mkdir("outcome") 
with open("outcome/hw_diamonds.csv", "w") as fw:
    fw.write("carat/cut/color/clarity/depth/table/price\n")      
    
    for idx, row in df.iterrows():
        fw.write("{}/{}/{}/{}/{}/{}/{}\n".format(row.carat, row.cut, row.color,row.clarity, row.depth, row.table, row.price))
```

```python
# Check your answer here. (Do not make any change to this cell. Just run this cell.)
os.path.isfile("outcome/hw_diamonds.csv")                # Expected output: True
```

2\. (1 pt) Read the file `hw_diamonds.csv` created in Question 1 and print the first 10 rows in the file including the header row, separated by a comma (,) between columns. For example, the output must be printed as below. Do not use Pandas.

```python
# Your answer here
with open("outcome/hw_diamonds.csv", "r") as fr:
    lines = fr.readlines()                 # Read the whole content in the file as a list of lines.
    
    # Decompose each line into values
    for line in lines[:10]:                 # Start from the second row.
        line = line.rstrip()               # Remove the trailing new line in each line.
        carat_val, cut_val, color_val, clarity_val, depth_val, table_val, price_val = line.split("/")
        print("{},{},{},{},{},{},{}".format(carat_val, cut_val, color_val, clarity_val, depth_val, table_val, price_val), end = "\n");
```

3\. (1 pt) Read the file `hw_diamonds.csv` created in Question 1 into a Pandas dataframe `df_new`. Make sure that the dataframe has 7 columns and 53,940 rows. You should be aware of the column delimiter used in the CSV file.

4\. (1 pt) Create a list `diamonds`, in which each element is a dictionary with the 7 keys being `carat`, `cut`, `color`, `clarity`, `depth`, `table`, and `price` and the values being their corresponding values in `df_new`. For example, the first element in `diamonds` must look as below. Then, serialize `diamonds` as a JSON formatted stream to a file named `hw_diamonds.json` saved in the `outcome` directory.

```python
# Your answer here
import json
diamonds = []
for idx,val in df_new.iterrows():
    item={"carat":val.carat,"cut":val.cut,"color":val.color,"clarity":val.clarity,"depth":val.depth,"table":val.table,"price":val.price}
    diamonds.append(item)
    
with open("outcome/hw_diamonds.json", "w") as fw:
    json.dump(diamonds, fw)
```
