<b>1.4 (3 pts)</b> You have an empty dictionary <i><b>counter</b></i>. Write a for loop that iterates the letters in <b><i>letters</i></b>. For each iteration, check if the current letter is already in <i><b>counter</b></i> as a key. If not, add to <i><b>counter</b></i> the letter as a new key and 1 as its value. If <i><b>counter</b></i> does have the letter as its key, increase the current value for the key by one. 

This way, you will have <i><b>counter</b></i> such that its keys are the unique letters in <b><i>letters</i></b> and its values are the occurrence counts of the keys. <u>For example</u>, the letter c appears three times in <b><i>letters</i></b>, so c is one of the keys in <i><b>counter</b></i> and its value is supposed to be 3. 

```python
# Do not make any change to this cell. Just run this cell.
counter = {}   # alterantively counter = dict()
```

```python
# Your answer here
for letter in letters:
    if letter in counter:
        counter[letter] +=1;
    else:
        counter[letter] = 1;
```

<b>2.1 (2 pts)</b> Write a function <b>get_counter</b> which takes a list of English alphabet strings <b>l</b> as an argument. E.g. <b>letters</b> from Q1 is a list with only English alphabet string elements.   l = ['z', 'c', 'x'] is another example of a list with only English alphabet strings as its elements. 

The output of get_counter must be a dictionary with keys being the unique alphabets in l and the value as the count of that alphabett in the list l. <u>For example</u>, the letter c appears three times in <b><i>letters</i></b>, so c must be one of the keys in the dictionary and its value is supposed to be 3. 

Essentailly we want the answer to Q1.4 written as a function. 

```python
# Your answer here
def get_counter(letters):
    counter = {};
    for letter in letters:
        if letter in counter:
            counter[letter] +=1;
        else:
            counter[letter] = 1;
    return counter;
```

* If there is no number in <b>l</b> less than <b>t</b>, it must return an empty list.  
* If any of the elements in <b>l</b> is neither an integer nor a floating pointer number, it returns the string "Invalid argument!"

Context: The list <b>l</b> consists of performance review scores out of 100. Those who score less than the satisfactory threshold <b>t</b> will be further evaluated. 

```python
# Your answer here
def further_check(l,t):
    th_result = [];
    for item in l:
        if (type(item) not in [int, float]):
                return 'Invalid argument!';
        else: 
            if(item < t):
                th_result.append(item);
    return th_result;
            
```

<b>3.3 (3 pts)</b> Write a function <b><i>sum_product</i></b> that takes two NumPy arrays <b><i>x</i></b> and <b><i>y</i></b> as arguments and returns the <u>sum</u> of the elements in the <u>dot product</u> of <b><i>x</i></b> and <b><i>y</i></b>. If the number of <u>columns</u> of <b><i>x</i></b> does <u>not</u> equal the number of rows of <b><i>y</i></b>, it returns the string "Array shapes not aligned!". 

```python
# Your answer here
def sum_product(x, y):
    xrow, xcolumn = x.shape
    yrow, ycolumn = y.shape
    if(xcolumn != yrow):
        return 'Array shapes not aligned!';
    dot_products = np.dot(x,y);
    return dot_products.sum();
    
```

```python
# Test your function here. (Do not make any change to this cell. Just run this cell.)
p = np.array([[1, 2], [3, 4]])
q = np.array([[6, 5], [4, 3], [2, 1]])

sum_product(p, q)                  # Expected output: 'Array shapes not aligned!'
```

<b>3.7 (2 pts)</b> We have been told that <b>probability</b> of getting heads (or tails) on a fair coin toss is 0.5. We would like to verify this claim uisng our Python skills. We will simulate a large number (n = 100000) of coin tosses in Python and count how many time we got the heads. 

We simluate a coin toss by randomly picking an integer from array [0, 1] and if it's 1, we count it as a heads. 

<b>Complete</b> the code given below to evaluate the <b>probability</b> of heads. 

```python
# complete and run this code to get the answer

np.random.seed(seed=0)
n = 100000    # number of coin tosses
count = 0    # to keep count of # of heads

for i in range(n):        # running the for loop n times
    # write down code to randomly chose an integer from an array [0,1]  and save it as toss
    toss = np.random.randint(0,2,1);
    # write down an if condition for checking if toss is 1 
    if(toss == 1):
        # write down code to update count by 1 if toss is 1
        count +=1;
        
Probability = count/n   # probability of heads 
Probability             # expect it to be close to 0.5
```
