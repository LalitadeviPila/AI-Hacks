2\. (0.5 pts) Add two new columns `words`, which contains the words identified from the `tweet` column, and `tagged_words`, which contains the part-of-speech tag information identified from the `words` column. Import any module necessary.

```python
# Your answer here
import nltk
dft["words"] = dft.tweet.apply(lambda x: nltk.word_tokenize(x))
dft["tagged_words"] = dft.words.apply(lambda x: nltk.pos_tag(x))
```

```python
# Check your answer here. (Do not make any change to this cell. Just run this cell.)
dft[["tweet", "words", "tagged_words"]]
```

3\. (0.5 pts) Add a new column `num_words` that contains the number of words in the `words` column.

```python
# Check your answer here. (Do not make any change to this cell. Just run this cell.)
dft[["tweet", "words", "num_words"]]
```

4\. (0.5 pts) Add two new columns `polarity` and `subjectivity` that contain the polarity score and subjectivity score of the `tweet` column, respectively. Import any module necessary. 

```python
# Your answer here
from textblob import TextBlob
dft["polarity"] = dft.tweet.apply(lambda x: TextBlob(x).sentiment.polarity)
dft["subjectivity"] = dft.tweet.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
# Check your answer here. (Do not make any change to this cell. Just run this cell.)
dft[["tweet", "polarity", "subjectivity"]]
```

(0.5 pts) Plot an interactive plot in which a radio buttons widget (refer to the documentation below) allows users to select a cut type among *All*, *Fair*, *Good*, *Ideal*, *Premium*, and *Very Good*, and interactively displays the dataframe that contains the rows with the selected cut type. For example, if the user selects the *Premium* cut type, it displays the dataframe that contains the rows with the Premium cut. If the user selects *All*, it displays the entire dataframe with all rows. The default value for the widget is *All*. To display a dataframe `dftmp`, use **display(dftmp)**. Do not use the **print** function

RadioButtons Widget: https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#RadioButtons

```python
# Your answer here
from ipywidgets import widgets, interactive, Layout
w_cuttype = widgets.RadioButtons(
    description = 'Cut type:',
    options =  ["All"] + sorted(set(dfdi.cut)),
    #['All', 'Fair', 'Good', 'Ideal', 'Preimum', 'Very Good'],
    style = {"description_width": '50px'},
    layout = Layout(width="70%"),
    disabled = False
)

def view(cut):
    if w_cuttype.value == "All":                         
        dftmp = dfdi                                   
    else:                                                
        dftmp = dfdi[dfdi.cut == w_cuttype.value]     

    title = "Cut type for {}".format(w_cuttype.value)
    dftmp[["carat", "depth", "table", "price", "x", "y", "z"]].plot(kind="box", title=title, grid=True, figsize=(15,7))
    
i = interactive(view, cut=w_cuttype)     # The value of the widget is passed to the view function as a parameter value.
display(i)
```
